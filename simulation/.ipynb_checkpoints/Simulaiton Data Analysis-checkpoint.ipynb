```python
import matplotlib
import matplotlib.pyplot as plt
from statistics import mean
import json
import numpy as np
# from scipy.stats.stats import pearsonr  
import os
import genome
import board as brd
import hyperParameters as hp
import seaborn as sns
from matplotlib.colors import ListedColormap

# construct cmap
flatui = [ "#FFFFFF", "#FFFFFF", "#000000", "#FF0000", "#00FF00"]
my_cmap = ListedColormap(sns.color_palette(flatui).as_hex())
```

```python
class SimulationRun:
  def __init__(self, filename):
    f = open(filename, "r")
    data = json.load(f)
    self.genomeData = data["genomes"]
#     self.lookupTableOutputs = lookupTableOutputs

    self.filename = filename
    self.numGens = len(self.genomeData)
    self.simulationData = []
    self.topFitnesses = []
    self.topSizeFitnesses = []
    self.topCellRatioFitnesses = []
    self.topAspectRatioFitnesses = []
    self.r1 = []
    self.r2 = []
    self.r3 = []
    self.finite = []
    self.infinite = []
    self.ranks = []

    self.topFitnesses = [d["fitness"] for d in self.genomeData]
    self.topCellRatioFitnesses = [d["cellRatioFitness"] for d in self.genomeData]
    self.topAspectRatioFitnesses = [d["aspectRatioFitness"] for d in self.genomeData]
    

    self.metR1 = [d["metabolicReservoirValues"]["0"] for d in self.genomeData]
    self.infoR1 = [d["informationalReservoirValues"]["0"] for d in self.genomeData]
    self.infoR2 = [d["informationalReservoirValues"]["1"] for d in self.genomeData]

    self.resHits = [d["resHits"] for d in self.genomeData]
    self.phase1s = [d[0] for d in self.resHits]
    self.phase2s = [d[1] for d in self.resHits]
    self.phase3s = [d[2] for d in self.resHits]
    self.phase4s = [d[3] for d in self.resHits]
    
    self.infiniteHits = []
    self.finiteHits = []
    
    for rh_g in self.resHits:
        sFin = sum([rh_g[0]["metFin"], rh_g[1]["metFin"], rh_g[2]["metFin"], rh_g[3]["metFin"]])
        sInf = sum([rh_g[0]["metInf"], rh_g[1]["metInf"], rh_g[2]["metInf"], rh_g[3]["metInf"]])
        self.finiteHits.append(sFin)
        self.infiniteHits.append(sInf)
    
    
    max_fitness = max(self.topFitnesses)
    self.bestGenomeIndex = self.topFitnesses.index(max_fitness)
    
    total = [i + f for (i, f) in zip(self.infinite, self.finite)]
    total = 1

    self.ranks = self.getTopFitnessInfo()

  def plotTopFitnesses(self):
    plt.plot(self.topFitnesses)
    plt.title(self.filename)
    plt.ylabel("Top Fitness")
    plt.xlabel("Generation Number")
    plt.grid(True)
    plt.show()

  def plotFitnessBreakdown(self):
    plt.plot(range(self.numGens), self.topSizeFitnesses, label="topSizeFitnesses")
    plt.plot(range(self.numGens), self.topCellRatioFitnesses, label="topCellRatioFitnesses")
    plt.plot(range(self.numGens), self.topAspectRatioFitnesses, label="topAspectRatioFitnesses")
    plt.title(self.filename)
    plt.ylabel("Fitness")
    plt.xlabel("Generation Number")
    plt.legend(labels=["top size fitness", "top cell ratio fitness", "top aspect ratio fitness"])
    plt.grid(True)
    plt.show()

  def plotMetabolicReservoirValues(self):
    plt.plot(range(self.numGens), self.metR1, label="r1")
    plt.title(self.filename)
    plt.ylabel("Initial Metabolic Reservoir Values")
    plt.xlabel("Generation Number")
    plt.legend(labels=["r1"])
    plt.grid(True)
    plt.show()

  def plotInformationalReservoirValues(self):
    plt.plot(range(self.numGens), self.infoR1, label="r1")
    plt.plot(range(self.numGens), self.infoR2, label="r2")
    plt.title(self.filename)
    plt.ylabel("Initial Informational Reservoir Values")
    plt.xlabel("Generation Number")
    plt.legend(labels=["r1", "r2"])
    plt.grid(True)
    plt.show()

  def plotFiniteVsInfiniteReservoirUsage(self):
    plt.plot(range(self.numGens), self.finite, label="finite")
    plt.plot(range(self.numGens), self.infinite, label="infinite")
    plt.title(self.filename)
    plt.ylabel("% Reservoir Usage")
    plt.xlabel("Generation Number")
    plt.legend(labels=["finite", "infinite"])
    plt.grid(True)
    plt.show()
  
  def plotAllDataFromRuns(self):
    fig, (ax1) = plt.subplots(1, 1, figsize=(30,5))
    
    fig.suptitle(self.filename)

    ax1.plot(self.topFitnesses)
    ax1.set(xlabel='Generation Number', ylabel='Top Fitness', title="Top Overall Fitness Score")
    ax1.grid(True)

#     ax2.plot(range(self.numGens), self.metR1, label="r1")
#     # ax3.plot(range(self.numGens), self.r2, label="r2")
#     # ax3.plot(range(self.numGens), self.r3, label="r3")
#     ax2.set(xlabel='Generation Number', ylabel='Initial Metabolic Reservoir Value', title="Initial Metabolic Reservoir Values")
#     ax2.legend(labels=["r1"])
#     ax2.grid(True)

#     ax3.plot(range(self.numGens), self.infiniteHits, label="Infinite")
#     ax3.plot(range(self.numGens), self.finiteHits, label="Finite")
#     # ax3.plot(range(self.numGens), self.r2, label="r2")
#     # ax3.plot(range(self.numGens), self.r3, label="r3")
#     ax3.set(xlabel='Generation Number', ylabel='Resevoir Hits', title="Reservoir Usage By Reservoir Type")
#     ax3.legend(labels=["Infinite", "Finite"])
#     ax3.grid(True)
    
#     ax4.plot(range(self.numGens), [d["metFin"] for d in self.phase1s], label="metabolic finite")
#     ax4.plot(range(self.numGens), [d["metInf"] for d in self.phase1s], label="metabolic infinite")
#     ax4.plot(range(self.numGens), [d["infFin"] for d in self.phase1s], label="informational finite")
#     ax4.plot(range(self.numGens), [d["infInf"] for d in self.phase1s], label="informational infinite")
#     ax4.set(xlabel='Generation Number', ylabel='Phase 1 Reservoir Usage', title="Phase 1 Reservoir Usage")
#     ax4.legend(labels=["metabolic finite", "metabolic infinite", "informational finite", "informational infinite"])
#     ax4.grid(True)

#     ax5.plot(range(self.numGens), [d["metFin"] for d in self.phase2s], label="metabolic finite")
#     ax5.plot(range(self.numGens), [d["metInf"] for d in self.phase2s], label="metabolic infinite")
#     ax5.plot(range(self.numGens), [d["infFin"] for d in self.phase2s], label="informational finite")
#     ax5.plot(range(self.numGens), [d["infInf"] for d in self.phase2s], label="informational infinite")
#     ax5.set(xlabel='Generation Number', ylabel='Phase 2 Reservoir Usage', title="Phase 2 Reservoir Usage")
#     ax5.legend(labels=["metabolic finite", "metabolic infinite", "informational finite", "informational infinite"])
#     ax5.grid(True)

#     ax6.plot(range(self.numGens), [d["metFin"] for d in self.phase3s], label="metabolic finite")
#     ax6.plot(range(self.numGens), [d["metInf"] for d in self.phase3s], label="metabolic infinite")
#     ax6.plot(range(self.numGens), [d["infFin"] for d in self.phase3s], label="informational finite")
#     ax6.plot(range(self.numGens), [d["infInf"] for d in self.phase3s], label="informational infinite")
#     ax6.set(xlabel='Generation Number', ylabel='Phase 3 Reservoir Usage', title="Phase 3 Reservoir Usage")
#     ax6.legend(labels=["metabolic finite", "metabolic infinite", "informational finite", "informational infinite"])
#     ax6.grid(True)

#     ax7.plot(range(self.numGens), [d["metFin"] for d in self.phase4s], label="metabolic finite")
#     ax7.plot(range(self.numGens), [d["metInf"] for d in self.phase4s], label="metabolic infinite")
#     ax7.plot(range(self.numGens), [d["infFin"] for d in self.phase4s], label="informational finite")
#     ax7.plot(range(self.numGens), [d["infInf"] for d in self.phase4s], label="informational infinite")
#     ax7.set(xlabel='Generation Number', ylabel='Phase 4 Reservoir Usage', title="Phase 4 Reservoir Usage")
#     ax7.legend(labels=["metabolic finite", "metabolic infinite", "informational finite", "informational infinite"])
#     ax7.grid(True)
    
    fig.savefig(self.filename + '.png')
    

  def getTopFitnessInfo(self):
    rs = []
    i = 0
    while (i < self.numGens and len(rs) < 10):
      if (self.topFitnesses[i] > len(rs) * 10):
        rs.append(i)
      else:
        i += 1
    return rs

  def plotRanks(self):
    plt.plot(self.ranks, range(len(self.ranks)))
    plt.title(self.filename)
    plt.ylabel("Passing fitnesses milestones (multiples of 10)")
    plt.xlabel("Generation Number")
    plt.grid(True)
    plt.show()

  def gensToGet50Fitness(self):
    for i, f in enumerate(self.topFitnesses):
      if f >= 50:
        return i
    return -1 
  
  def gensToGet90Fitness(self):
    for i, f in enumerate(self.topFitnesses):
      if f >= 90:
        return i
    return -1 

  def beforeAndAfterDeletion(self):
        
        deleteCellSpecification = {
            "N": 45,
            "S": 55,
            "E": 60,
            "W": 45
        }
        
        def convertStringKeysToIntKeys(d):
            keys = list(d.keys()) 
            for k in keys:
                d[int(k)] = d[k]
                d.pop(k, None)
            return d
        
        board = brd.Board(hp.boardWidth, hp.boardHeight)
        
        gi = self.genomeData[self.bestGenomeIndex]
        g = genome.Genome(gi["genome"], 
            convertStringKeysToIntKeys(gi["metabolicReservoirValues"]), 
            convertStringKeysToIntKeys(gi["informationalReservoirValues"]), 
            gi["fitness"],  
        )
        g.fillReservoirs()
        board.reset(g)
        while (len(board.dynamicCells)):
            board.step()
            
        fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(30,5))

        fig.suptitle(self.filename)

        data = np.array(board.grid)

        rows,cols = data.shape
                
        
        ax1.imshow(data, interpolation='none', 
                        extent=[0.5, 0.5+cols, 0.5, 0.5+rows], 
                        aspect="equal",
                        cmap=my_cmap)
        
        for i in range(board.height):
            for j in range(board.width):
                if j < deleteCellSpecification["E"] and j > deleteCellSpecification["W"] and i > deleteCellSpecification["N"] and i < deleteCellSpecification["S"]:
                    board.grid[i][j] = 0
                    
        data = np.array(board.grid)

        rows,cols = data.shape

        ax2.imshow(data, interpolation='none', 
                    extent=[0.5, 0.5+cols, 0.5, 0.5+rows], 
                    aspect="equal",
                    cmap=my_cmap)

        for c in board.staticCells:
            board.dynamicCells.add(c)

        while (len(board.dynamicCells)):
            board.step()

        data = np.array(board.grid)

        rows,cols = data.shape

        ax3.imshow(data, interpolation='none', 
                    extent=[0.5, 0.5+cols, 0.5, 0.5+rows], 
                    aspect="equal",
                    cmap=my_cmap)

        
        fig.savefig(self.filename + "_figure" +'.png')

  def showBestAgent(self):
        def convertStringKeysToIntKeys(d):
            keys = list(d.keys()) 
            for k in keys:
                d[int(k)] = d[k]
                d.pop(k, None)
            return d
        
        board = brd.Board(hp.boardWidth, hp.boardHeight)
        
        gi = self.genomeData[self.bestGenomeIndex]
        g = genome.Genome(gi["genome"], 
            convertStringKeysToIntKeys(gi["metabolicReservoirValues"]), 
            convertStringKeysToIntKeys(gi["informationalReservoirValues"]), 
            gi["fitness"],  
        )
        
        
        for i in range(10):
            g.fillReservoirs()
            board.reset(g)
            while (len(board.dynamicCells)):
                board.step()

            data = np.array(board.grid)

            rows,cols = data.shape

            plt.imshow(data, interpolation='none', 
                            extent=[0.5, 0.5+cols, 0.5, 0.5+rows], 
                            aspect="equal",
                            cmap=my_cmap)

            plt.savefig(self.filename + "_repeat_" + str(i) + "_img" + '.png')
    

  def getLookupTableOutputs(self):
        def convertStringKeysToIntKeys(d):
            keys = list(d.keys()) 
            for k in keys:
                d[int(k)] = d[k]
                d.pop(k, None)
            return d
        
        board = brd.Board(hp.boardWidth, hp.boardHeight)
        
        gi = self.genomeData[self.bestGenomeIndex]
        g = genome.Genome(gi["genome"], 
            convertStringKeysToIntKeys(gi["metabolicReservoirValues"]), 
            convertStringKeysToIntKeys(gi["informationalReservoirValues"]), 
            gi["fitness"],  
        )
        g.fillReservoirs()
        board.reset(g)
        while (len(board.dynamicCells)):
            board.step()
            
        _, _, filenames = next(os.walk("."), (None, None, []))
        print(filenames)

        f = open('/lookup_table_outputs_temp.txt', 'r')
        Lines = f.readlines()

        d = {}
        for el in Lines:
            try:
                d[str(el)] += 1
            except:
                d[str(el)] = 1
                
        with open(self.filename + ".txt", 'a+') as outfile:
            print("lookup table length: ", len(runsA[0].genomeData), file=outfile)
            print("number of outputs used: ", len(Lines), file=outfile)
            print("unique outputs used: ", len(d), file=outfile)
            print("-----------", file=outfile)

            for w in sorted(d, key=d.get, reverse=True)[:6]:
                print("Genome output: ", w, file=outfile)
                print("Number of times used: ", d[w], file=outfile)
                print("----", file=outfile)
            
            outfile.close()                  
```
